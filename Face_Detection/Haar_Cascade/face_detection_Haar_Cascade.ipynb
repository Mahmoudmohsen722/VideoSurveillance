```python
def face_detect_pipeline(img, scale_factor=1.1, min_Neighbors=3):
    gray = cv.cvtColor(img, cv.COLOR_RGB2GRAY)
    haar_cascade = cv.CascadeClassifier('./Resources/haar_face.xml')
    faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor= scale_factor, minNeighbors= min_Neighbors)
    for (x,y,w,z) in faces_rect:
        cv.rectangle(img, (x, y), (x+w, y+z), (0,0,255), 2)
```

```python
img2 = cv.imread('./Resources/Photos/group 1.jpg')
```

```python
face_detect_pipeline(img2)
show_image(' ', img2)
plt.imshow(img2)
```

```python
def face_detect_video(input_path, scale_factor=1.1, min_Neighbors=5):
    cap = cv.VideoCapture(input_path)
    while True:
        _ , frame = cap.read()
        gray = cv.cvtColor(frame, cv.COLOR_RGB2GRAY)
        haar_cascade = cv.CascadeClassifier('./Resources/haar_face.xml')
        faces_rect = haar_cascade.detectMultiScale(gray, scaleFactor= scale_factor, minNeighbors= min_Neighbors)
        for (x,y,w,z) in faces_rect:
            cv.rectangle(img, (x, y), (x+w, y+z), (0,0,255), 2)
            
        cv.imshow('frame', frame)
        if cv.waitKey(1) & 0xFF == ord('q'):
            break
            
    cap.release()
# Destroy all the windows
    cv.destroyAllWindows()
```

```python
face_detect_video('./M G ROAD CROWD WALKING _ STOCK FOOTAGES.mp4')
```
